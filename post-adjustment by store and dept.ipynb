```python
myeva_new = pd.DataFrame()
adj_range = np.linspace(0.1,3,300)

for k in range(0,10):
    myeva_full = pd.read_excel(f'{DATA_DIR}/fold_{k+1}/myvalidate.xlsx')
    for i in range(1,100):
    
        print(f'Dept {i} improvement start')
    
        myeva = myeva_full[myeva_full['Dept']==i]
        myeva_temp = myeva.copy()
        weights = myeva['weights']
        pred = myeva['Weekly_Pred']
        actual = myeva['Weekly_Sales']
        current_adj=1
        current_wae = (abs(pred*current_adj-actual)*weights).sum()/weights.sum()
        print(f'initial WAE is {current_wae}')
        
        for adj in adj_range:
            adj = round(adj,2)
            new_adj = adj
            new_wae = (abs(pred*new_adj-actual)*weights).sum()/weights.sum()
    
            if new_wae < current_wae:
                print(f'a better adjustment for Dept {i} is found: adjust is {adj} with WAE = {new_wae}')
                current_adj = adj
                current_wae = new_wae
        myeva_temp['folder'] = k+1
        myeva_temp['adj'] = current_adj
        myeva_temp['new_error'] = myeva['Weekly_Sales'] - (myeva['Weekly_Pred']*current_adj)
        myeva_temp['new_abs_error'] = abs(myeva_temp['new_error'])
        myeva_temp['new_weighted_abs_error'] =  myeva_temp['new_abs_error']*myeva_temp['weights']
    
        myeva_new = pd.concat([myeva_new, myeva_temp], axis=0, ignore_index=True)
        overall_wae = (abs(myeva_new['Weekly_Pred']*myeva_new['adj']-myeva_new['Weekly_Sales'])*myeva_new['weights']).sum()/myeva_new['weights'].sum()
        print(f'current overall WAE is:{overall_wae}')

myeva_new.to_csv(f'{DATA_DIR}/myeva_new_by dept.csv')
```
